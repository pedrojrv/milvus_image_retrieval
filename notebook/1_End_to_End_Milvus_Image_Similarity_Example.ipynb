# Setting Up the ResNet50 Model for Embedding Generation

The `image_to_vec.ImageVectorizer()` class loads the `ResNet50` model and removes the last couple of layers so that we can get the last average pooling layer output. This output is our embeddings and are equivalent to our deep features. The dimensions of the vector generated by this layer is `2048`.

```python
# we first instantiate our Image Vectorizer class
embedding_generator = image_to_vec.ImgageVectorizer()
```

```python
# Let us get the embedding for one of the sample images provided
example_vector = embedding_generator.get_embedding("../data/example_images/20200327_134637.jpg")
example_vector
```

```python
example_vector.reshape(1, -1).shape
```

```python
# The shape in our generated embedding should be 2048
assert example_vector.shape[0] == 2048
example_vector.shape
```

# Generating Embeddings for an Entire Directory

To accelerate the process of generating embeddings for directories of images the `.get_embeddings_from_directory()` function is provided. Let us see how it works in the `example_images` directory provided which contains 10 images.

```python
image_directory = "../data/example_images/"
image_extension = ".jpg"

all_embeddings, filepaths = image_to_vec.get_embeddings_from_directory(
    image_directory, extension=image_extension)
```

```python
# we should have 10 examples each with 2048 features
assert all_embeddings.shape == (10, 2048)
```

The function returns two objects: a numpy array containing all generated embeddings, and the path to all image files so we can keeep track of them for later. This information can then be stored in a database of your choice. For this simple case we shall keep it local and save the embeddings and metadata information into a `csv` file.

```python
import pandas as pd
```

```python
df_embeddings = pd.DataFrame(all_embeddings)
# append the image_paths to our csv files
df_embeddings['image_path'] = filepaths
# we save the file in the data directory
df_embeddings.to_csv("../data/example_embeddings.csv", index=False)
```

## Loading the Downloaded File and Getting the ImageNet Embedding Array

```python
imagenet_embeddings = pd.read_csv("../data/imagenet_vectors.csv")
```

```python
# append the data path to the image_path feature
imagenet_embeddings['image_path'] = imagenet_embeddings['image_path'].apply(
    lambda x: os.path.join("../data/", x))
```

```python
# We extract the embedding as a numpy array to upload them to milvus
embedding_vectors = imagenet_embeddings.iloc[:,:-1].values
```

# Creating a Milvus Collection and Uploading our Embeddings

Before proceeding, make sure you pulled the milvus docker container and that is up and running according to the `README.md` file of this repository. You can confirm by running `docker ps` in the terminal:

```python
!docker ps
```

In the above output you can observe that the `port` is `19530`. Let us connect to the `milvus` server.

```python
from milvus import Milvus, IndexType, MetricType, Status
```

```python
# change the HOST and PORT if applicable
_HOST = '127.0.0.1'
_PORT = '19530' 

# We will need the dimension of our vector information to create a collection
_DIM = embedding_vectors.shape[1]  

collection_name = 'imagenet_collection'

milvus = Milvus(_HOST, _PORT)
```

```python
import importlib
importlib.reload(milvus_utilities)
```

```python
# This utility function creates a collection or warns if it already exists
milvus_utilities.create_collection(milvus, collection_name, _DIM, reset=True)
```

Now that our collection is up and running, we can upload the ImageNet embeddings you just downloaded. The `insert_embeddings()` function makes this easy. In case you did not notice, there is a `buffer_size` specified in the `server_config.yaml` file when running the milvus server. The default is `256MB` but the imagenet embeddigngs are heavier:

```python
print("ImageNet Embeddings Size: {:.0f}MB".format(embedding_vectors.nbytes * 1e-6))
```

The `insert_embeddings()` function detects the size of your array and inserts the embeddings in chunks to avoid errors raised by the `milvus` server.

```python
inserted_ids = milvus_utilities.insert_embeddings(milvus, collection_name, embedding_vectors, buffer_size=256)
```

```python
# we can make sure all our vectors were indeed uploaded by asserttion
# if you get an error try running the cell again, it takes a couple of seconds for the collection to update
status, result = milvus.count_entities(collection_name)
assert embedding_vectors.shape[0] == result
```

```python
milvus.flush([collection_name])
```

The function returns the `ids` for the uploaded vectors. Remember our original `csv` file containing the image paths? Well, we can append the returned `ids` to the `csv` file to keep track of the corresponding image of each `id`. 

```python
imagenet_embeddings["milvus_ids"] = inserted_ids
```

```python
imagenet_embeddings.to_csv("../data/imagenet_vectors_w_ids.csv", index=False)
```
